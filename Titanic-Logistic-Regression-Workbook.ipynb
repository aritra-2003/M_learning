```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

## Data Loading

```python
data = pd.read_csv("Titanic-Train-Data.csv")
```

```python
data.head()
```

```python
data.describe()
```

## Data Preprocessing and Cleaning

```python
data.drop(["Name","Ticket"],axis=1,inplace=True)
```

```python
def male_ohe(g):
    if g == "male":
        return 1
    else:
        return 0

def female_ohe(g):
    if g == "female":
        return 1
    else:
        return 0
```

```python
data["Male"] = data["Sex"].apply(male_ohe)
data["Female"] = data["Sex"].apply(female_ohe)
```

```python
data.drop("Sex",axis=1,inplace=True)
data.head()
```

```python
data.isnull().sum()
```

```python
def cherbourg(s):
    if s=="C":
        return 1
    else:
        return 0
    
def queenstown(s):
    if s=="Q":
        return 1
    else:
        return 0

def southamptom(s):
    if s=="S":
        return 1
    else:
        return 0
```

```python
data["Cherbourg"] = data.Embarked.apply(cherbourg)
data["Queenstown"] = data.Embarked.apply(queenstown)
data["Southamptom"] = data.Embarked.apply(southamptom)
data.drop("Embarked",axis=1, inplace=True)
data.head()
```

```python
def cabin(x):
    if x is np.nan:
        return 0
    else:
        return 1
```

```python
data["Cabin"] = data.Cabin.apply(cabin)
data.head()
```

```python
data.isnull().sum()
```

```python
data["Age"].fillna(data["Age"].mean(),inplace=True)
data.head()
```

```python
data.describe()
```

```python
data = data[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare','Cabin','Male','Female','Cherbourg','Queenstown','Southamptom','Survived']]
data.head()
```

## To Numpy

```python
data = data.values
```

```python
data.shape
```

```python
X = data[:,0:11]
Y = data[:,11]
X.shape,Y.shape
```

## Feature Scaling

```python
scaler = preprocessing.StandardScaler()
```

```python
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
X_scaled.shape
```

```python
X_scaled = np.append(X_scaled,np.ones(X_scaled.shape[0]).reshape(-1,1),axis=1)
```

```python
X_scaled.shape,X_scaled[0].shape
```

## Train Code

```python
def sigmoid(agg):
    
    return 1/(1+np.exp(-agg))

def cost(X_train,Y_train,m):
    
    cost_ = 0
    N = X_train.shape[0]
    for i in range(N):
        agg = (X_train[i]*m).sum()
        h = sigmoid(agg)
        cost = -Y_train[i]*np.log(h) - (1-Y_train[i])*np.log(1-h)
        cost_ += cost
    
    return cost_

def step_gradient(X_train,Y_train,lr,m):
    
    N = X_train.shape[0]
    slope_m = np.zeros(X_train.shape[1])
    for i in range(N):
        agg = (X_train[i]*m).sum()
        h = sigmoid(agg)
        slope_m+=(-1/N)*(Y_train[i]-h)*X_train[i]
        
    m = m - lr*slope_m
    return m

def fit(X_train,Y_train,epochs=100,lr=0.01):
    
    m = np.zeros(X_train.shape[1])
    cost_array = []
    unit = epochs//10
    for i in range(epochs+1):
        m = step_gradient(X_train,Y_train,lr,m)
        cost_ = cost(X_train,Y_train,m)
        cost_array.append(cost_)
        if i%unit==0:
            print("Epoch:{}, Cost:{}".format(i,cost_))
    
    return m,cost_array

def predict(X_test,m):
    
    y_pred = []
    N = X_test.shape[0]
    for i in range(N):
        agg = (X_test[i]*m).sum()
        h = sigmoid(agg)
        if h>=0.5:
            y_pred.append(1)
        else:
            y_pred.append(0)
            
    return np.array(y_pred)

def accuracy(Y_test,Y_pred):
    
    correct = 0
    N = Y_test.shape[0]
    correct = (Y_test==Y_pred).sum()
    
    return (correct/N)*100
```

```python
m,cost_array = fit(X_scaled,Y,3000,1)
print(m)
```

```python
plt.plot(cost_array)
plt.grid()
plt.show()
```

```python
y_pred_train = predict(X_scaled,m)
```

```python
accuracy(Y,y_pred_train)
```
